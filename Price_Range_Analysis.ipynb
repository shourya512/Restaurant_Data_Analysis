#Task 2: Price Range Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##Load dataset

```python
df = pd.read_csv("/content/Dataset .csv")
df.head()
```

## Most Common Price Range

Identifying which price range appears most frequently in the dataset.


```python
df['Price range'].value_counts()
```

```python
df['Price range'].value_counts().plot(kind='bar', figsize=(6,4))
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.title("Distribution of Price Ranges")
plt.show()
```

Price range 1 is the most common among the restaurants.

## Average Rating by Price Range

Calculating the average aggregate rating for each price range.


```python
avg_rating_price = df.groupby('Price range')['Aggregate rating'].mean()
avg_rating_price
```

```python
avg_rating_price.plot(kind='bar', figsize=(6,4))
plt.xlabel("Price Range")
plt.ylabel("Average Rating")
plt.title("Average Rating by Price Range")
plt.show()
```

Higher price ranges generally show slightly higher average ratings.

## Rating Color vs Average Rating

Identifying which rating color represents the highest average rating.


```python
 avg_rating_color = df.groupby('Rating color')['Aggregate rating'].mean()
avg_rating_color
```

```python
avg_rating_color.plot(kind='bar', figsize=(6,4))
plt.xlabel("Rating Color")
plt.ylabel("Average Rating")
plt.title("Average Rating by Rating Color")
plt.show()
```

The color representing the highest average rating corresponds to higher-rated restaurants.
This color is typically associated with better customer feedback.


## Conclusion

Price range 1 is the most common among restaurants.
Average ratings tend to increase with higher price ranges.
Rating colors help visually represent restaurant quality,
with certain colors indicating higher average ratings.
