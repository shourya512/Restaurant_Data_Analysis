# Task 3: Geospatial Analysis
This task will visualize restaurant locations using it's latitude and longitude, analyzing restaurant distribution across cities and countries,
and checking if location has any relation with ratings.

##Imort libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##Loading dataset

```python
df = pd.read_csv("/content/Dataset .csv")
df.head()
```

##Cheching location
Geospatial analysis requires accurate latitude and longitude coordinates.

```python
df[['Latitude', 'Longitude']].describe()
```

The dataset includes latitude and longitude coordinates suitable for mapping restaurant locations.

## Restaurant Locations Using Latitude and Longitude


```python
plt.figure(figsize=(8,6))
plt.scatter(df['Longitude'], df['Latitude'], alpha=0.3)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Restaurant Locations")
plt.show()
```

Many restaurants are in the same places instead of everywhere.

## Distribution of Restaurants Across Countries


```python
df['Country Code'].value_counts().head(10)
```

```python
df['Country Code'].value_counts().head(10).plot(kind='bar', figsize=(8,4))
plt.xlabel("Country Code")
plt.ylabel("Number of Restaurants")
plt.title("Top Countries by Restaurant Count")
plt.show()
```

As per the observation Only some countries have most of the restaurants.

## Distribution of Restaurants Across Cities


```python
df['City'].value_counts().head(10)
```

```python
df['City'].value_counts().head(10).plot(kind='bar', figsize=(10,4))
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.title("Top 10 Cities by Restaurant Count")
plt.show()
```

As per the observation only a small number of cities have many restaurants.

## Relationship Between Location and Rating


```python
df[['Latitude', 'Longitude', 'Aggregate rating']].corr()
```

Longitude Vs Aggregate Rating.

```python
plt.figure(figsize=(6,4))
plt.scatter(df['Latitude'], df['Aggregate rating'], alpha=0.3)
plt.xlabel("Latitude")
plt.ylabel("Aggregate Rating")
plt.title("Latitude vs Aggregate Rating")
plt.show()
```

Longitude Vs Aggregate Rating.

```python
plt.figure(figsize=(6,4))
plt.scatter(df['Longitude'], df['Aggregate rating'], alpha=0.3)
plt.xlabel("Longitude")
plt.ylabel("Aggregate Rating")
plt.title("Longitude vs Aggregate Rating")
plt.show()
```

As per the observation there is no visible relation found between location and rating.
Ratings seem unrelated to geographic location.

## Conclusion

Restaurant locations were mapped using latitude and longitude. Most restaurants are found in a few cities and countries. The analysis shows that ratings are not strongly related to location.
