#Task 3: Feature Engineering
This task will focus on creating new features from existing columns to improve understanding and preparing the dataset for modeling.

##Import libraries

```python
import pandas as pd
```

## Load datset

```python
df = pd.read_csv("/content/Dataset .csv")
df.head()
```

## Length of Restaurant Name

Creating a new feature based on the number of characters
in the restaurant name.


```python
df['Restaurant_Name_Length'] = df['Restaurant Name'].str.len()
df[['Restaurant Name', 'Restaurant_Name_Length']].head()
```

As per the observation length of restaurants name varies and can be useful for analysis.

## Length of Restaurant Address

Creating a feature based on the length of the restaurant address.


```python
df['Address_Length'] = df['Address'].str.len()
df[['Address', 'Address_Length']].head()
```

As per the observation the address length provides information about how detailed the location is.

## Encoding Table Booking Feature

```python
df['Table_Booking_Flag'] = df['Has Table booking'].map({'Yes': 1, 'No': 0})
df[['Has Table booking', 'Table_Booking_Flag']].head()
```

As per the observation table booking availability is now represented as a binary feature.

## Encoding Online Delivery Feature


```python
df['Online_Delivery_Flag'] = df['Has Online delivery'].map({'Yes': 1, 'No': 0})
df[['Has Online delivery', 'Online_Delivery_Flag']].head()
```

As per the observation online delivery availability has been converted into a numeric format.

## Checking Newly Created Features

```python
df[['Restaurant_Name_Length', 'Address_Length',
    'Table_Booking_Flag', 'Online_Delivery_Flag']].describe()
```

## Conclusion

New features were successfully created from existing column.
Text-based features like name and address length were excited.
Categorical variables for table booking and online delivery were encoded into numeric format for further analysis or modeling.
