#Task 1: Table Booking and Online Delivery

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##Load dataset

```python
df = pd.read_csv("/content/Dataset .csv")
df.head()
```

## Percentage of Restaurants Offering Table Booking


```python
table_booking_counts = df['Has Table booking'].value_counts()
table_booking_percentage = (table_booking_counts / len(df)) * 100

table_booking_percentage
```

Only small percentage of restaurants offer table booking facilities.

## Percentage of Restaurants Offering Online Delivery


```python
online_delivery_counts = df['Has Online delivery'].value_counts()
online_delivery_percentage = (online_delivery_counts / len(df)) * 100

online_delivery_percentage
```

As per the observation the online delivery is available in a limited number of restaurants compared to those without it.

## Average Rating Comparison (Table Booking)

```python
avg_rating_table_booking = df.groupby('Has Table booking')['Aggregate rating'].mean()
avg_rating_table_booking
```

```python
avg_rating_table_booking.plot(kind='bar', figsize=(6,4))
plt.xlabel("Table Booking Availability")
plt.ylabel("Average Rating")
plt.title("Average Rating vs Table Booking")
plt.show()
```

As per the observation restaurants offering table booking tend to have slightly higher average ratings.

## Online Delivery Availability Across Price Ranges

```python
price_delivery = df.groupby('Price range')['Has Online delivery'].value_counts(normalize=True) * 100
price_delivery
```

```python
####To make it more easily readable:

price_delivery.unstack()
```

```python
price_delivery.unstack().plot(kind='bar', figsize=(8,4))
plt.xlabel("Price Range")
plt.ylabel("Percentage")
plt.title("Online Delivery Availability by Price Range")
plt.show()
```

As per the observation the Mid-range restaurants are offering more online delivery as compared to low and high price ranges.

## Conclusion

Only a small percentage of restaurants provide table booking and online delivery services.
Restaurants with table booking generally have slightly higher average ratings.
Online delivery availability varies across price ranges and is more common in mid-range restaurants.
