##Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

##Load Dataset

```python
df = pd.read_csv("/content/Dataset .csv")
df.head()
```

 ## Basic Statistical Measures of Numerical Columns
Calculating Mean, Median and Standard deviation

```python
df.describe()
```

The mean and median values shows that most of the rating lies between 3 and 4 and some columns have higher standard deviation which shows more variation.

## Distribution of Categorical Variables

###Country Code

```python
df['Country Code'].value_counts().head()
```

```python
df['Country Code'].value_counts().plot(kind='bar', figsize=(8,4))
plt.xlabel("Country Code")
plt.ylabel("Number of Restaurants")
plt.title("Distribution of Country Code")
plt.show()
```

Some country codes have higher number of restaurants compared to others.

### City

```python
df['City'].value_counts().head(10)
```

```python
df['City'].value_counts().head(10).plot(kind='bar', figsize=(10,4))
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.title("Top 10 Cities by Number of Restaurants")
plt.show()
```

A small number of cities have large number of restaurants

### Cuisines

```python
df['Cuisines'].value_counts().head(10)
```

```python
df['Cuisines'].value_counts().head(10).plot(kind='bar', figsize=(10,4))
plt.xlabel("Cuisine Type")
plt.ylabel("Number of Restaurants")
plt.title("Top 10 Cuisines")
plt.show()
```

Some cuisines show up much more often than others.
