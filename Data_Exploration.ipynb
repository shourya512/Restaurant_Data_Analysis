##Importing required Libraries

Using basic python libraries for data analysis and visualisation.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading the Dataset

```python
df = pd.read_csv('/content/Dataset .csv')
df.head()
```

## Dataset shape

Checking number of rows and columns in the dataset.

```python
df.shape
```

## Data types

```python
df.info()
```

It seems like most of the columns have correct data types and there is no conversion is required.

##Missing values check
Now, checking for the missing values in each column.

```python
df.isnull().sum()
```

Only the "cuisines" column contains missing values which is 9.

##Handle Missing values


```python
df['Cuisines'] = df['Cuisines'].fillna("Not Specified")
```

Missing cuisine values are replaced with 'Not Specified'.

##Target variable analysis
Analyzing the distribution of the target variable.

```python
df['Aggregate rating'].value_counts()
```

## Distribution of Aggregate Rating

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Aggregate rating'], bins=20, kde=True)
plt.xlabel("Aggregate Rating")
plt.ylabel("Count")
plt.title("Aggregate Rating Distribution")
plt.show()
```

Here is the graphical representation of Aggregate Rating Distribution.
